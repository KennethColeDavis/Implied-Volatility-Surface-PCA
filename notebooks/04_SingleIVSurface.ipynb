## Building a single Implied Volatility Surface with Yfinance Library
Steps:
1) Pull option chain for S&P 500 
2) Sort and filter data 
3) Calculate IV for each option 
4) Plot IV surface
5) Smooth IV surface

```python
#Pull option chain data 
import yfinance as yf
TickerName = 'SPY'
tk = yf.Ticker(TickerName)

expirations = tk.options
print("There are", len(expirations), "expirations")
print("The expirations are:", expirations)
print("The latest expiration is:", expirations[-1])
print("The collumns in the option chain data are:", tk.option_chain(expirations[0]).calls.columns)
```
